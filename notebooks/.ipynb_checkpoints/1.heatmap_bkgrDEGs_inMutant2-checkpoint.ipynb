Which genes change their expression when growth is induced by a temperature shift in the Pad4 mutant, i.e. a mutant without SA signalling?


T0 = 22°C
T1 = 1 day at 28°C
T2 = 4 days at 28°C

We made Venn diagrams in 0.Venn_diagrams_timepoints_mutants.


```python
# specific imports for making Venn diagrams
from matplotlib_venn import venn3, venn3_circles
from matplotlib_venn import venn2, venn2_circles
from matplotlib import pyplot as plt
%matplotlib inline

import pickle

maxQ = 0.01
S = 'pad4'
timepoints  = ['T0','T1','T2']
dirnameTime = 'Pad4_NEW/reanalysisTime/'

# make dict with probeset_id --> transcript_id mapping
def probe2transcripts(fname):
    p2t   = {}
    lines = open(fname).readlines()[1:]
    for line in lines:
        data        = line.strip().split('\t')
        probe       = int(data[0])
        transcripts = data[1].split(';')

        p2t[probe] = transcripts
        
    if not len(p2t.keys()) == len(lines):
        print 'WARNING: keys in probe2transcript dict not the number of lines in the file'
    
    return p2t

# select probes from file
def select_probes(fname, qvalue_cutoff):
    selection = set([])
    
    #fileheader:
    #probeset_id     transcript_id   logFC   AveExpr t       P.Value adj.P.Val       B       qvalues
    for line in open(fname).readlines()[1:]: #skip header
        data = line.strip().split('\t')
        if float(data[-1]) <= qvalue_cutoff:
            selection.add(int(data[0]))
            
    return selection

# Now, for each strain, the overlap in DEGs per timepoint comparison
probesets = []
labels    = []
allDEGs   = set([])
p2t       = probe2transcripts('Pad4_NEW/probeset_id2transcript_id.tab')
for x, Tx in enumerate(timepoints[:-1]):
    for y, Ty in enumerate(timepoints[x+1:]):

        # select DEGs in strains when comparing timepoint Tx to Ty
        probes = select_probes(dirnameTime+S+'_'+Tx+'_vs_'+Ty+'.txt', maxQ)
        
        # write transcripts to a file 
        data_fname      = 'Pad4_NEW/DEGs_in_Pad4/'+Tx+'-'+Ty+'.txt'
        transcript_file = open(data_fname, 'w')
        
        for p in probes:
            if p2t.has_key(p):
                for t in p2t[p]:
                    transcript_file.write(t+'\n')
            else:
                print 'Probe not found', p 
        
        transcript_file.close()
        
        pickle.dump(probes, 'Pad4_NEW/DEGs_in_Pad4/pickles/'+Tx+'-'+Ty+'.probes_as_set', 'w')
        probesets.append(probes)
        allDEGs = allDEGs.union(probes)
        labels.append(Tx+'-'+Ty+'\n('+str(len(probes))+')')  

pickle.dump(allDEGs, open('Pad4_NEW/DEGs_in_Pad4/pickles/ALL.probes_as_set', 'w')        
plt.figure()
plt.title(S+': '+str(len(allDEGs)))
venn = venn3(probesets, set_labels = labels)

```

Expression values are log transformed, taking the difference of two log-transformed variables is equivalent of taking the log of their ratio: log(X/Y)=log(X)−log(Y)



```python
# add extra columns to dataframe with ratio mutant/bkgr:
BKGR = strains[0]
for S in strains[1:]:
    diff_data = pd.DataFrame({'diff_T0_'+S+'_vs_'+BKGR: DEGs['T0_'+S]-DEGs['T0_'+BKGR],\
                           'diff_T1_'+S+'_vs_'+BKGR: DEGs['T1_'+S]-DEGs['T1_'+BKGR],\
                           'diff_T2_'+S+'_vs_'+BKGR: DEGs['T2_'+S]-DEGs['T2_'+BKGR]})


    #Now let's make a heatmap of the differences:
    sns.clustermap(diff_data, row_linkage=row_linkage, col_cluster=False, cmap = 'BrBG',vmin=-3,vmax=3)
    sns.plt.title('log('+S+'/'+BKGR+')')
    #sns.plt.savefig('Pad4_NEW/HEATMAP_DIFF_DEGs_'+S+'.png', dpi=700)
    sns.plt.savefig('Pad4_NEW/heatmaps/selection/HEATMAP_DIFF_DEGs_'+S+'.eps')
    sns.plt.show()
    
    #sns.clustermap(diff_data, row_linkage=row_linkage, col_cluster=False, cmap = 'RdBu_r')#,vmin=-4,vmax=4)
    #sns.plt.title('log('+S+'/'+BKGR+')')
    #sns.plt.savefig('Pad4_NEW/HEATMAP_DIFF_DEGs_'+S+'_RdBu.png', dpi=700)
    #sns.plt.show()
    
    
```
